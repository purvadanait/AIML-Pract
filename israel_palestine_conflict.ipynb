```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Admin\Downloads\archive\fatalities_isr_pse_conflict.csv")
df.head()
```

```python
sns.histplot(data=df, x='age', y='number', kde=True, hue='gender')
plt.title('Age Distribution by Gender')
plt.show()
```

```python
sns.barplot(data=df, x='citizenship', y='number', hue='gender')
plt.title('Fatalities')
plt.show()
```

```python
plt.plot(df['date'], df['number'])
plt.title('Fatalities by year')
plt.show()
```

```python
nation_grouped = df.groupby('citizenship')['nation'].count().reset_index()

plt.pie(nation_grouped['nation'], labels=nation_grouped['citizenship'])
plt.title('Number of fatalities by Nation')
plt.show()
```

```python
sns.barplot(data=df, x='killed_by', y='number', hue='gender')
plt.title('Killed by')
plt.show()
```

```python
sns.barplot(data=df, y='type_of_injury', x='number')
plt.title('Death by')
plt.show()
```
